**Objetivo:** Construir la tabla top 10 coloreada, el KPI de riesgo y la alerta condicional integrados en la pestana 'Riesgo de rotacion'.

```python
import pandas as pd
import numpy as np
import plotly.express as px

df = pd.read_csv('../hr_ibm.csv')

# Reutilizar calcular_score() de la practica C2

def calcular_score(df):
    d = df.copy()
    d['n_overtime'] = (d['OverTime'] == 'Yes').astype(int)*100
    d['n_jobsat'] = (4 - d['JobSatisfaction'])/3 *100
    d['n_years'] = ((3-d['YearsAtCompany'])/3*100).clip(0,100)
    d['n_wlb'] = (4 - d['WorkLifeBalance']) /3  *100
    d['n_envsat'] = (4 - d['EnvironmentSatisfaction'])/3 * 100

    d['score_riesgo'] = (
        d['n_overtime']*0.30 + d['n_jobsat']*0.25 +
        d['n_years']*0.20 + d['n_wlb']*0.15 +
        d['n_envsat']*0.10
    ).round(1)

    d['nivel_riesgo'] = pd.cut(d['score_riesgo'], bins=[0, 39, 69, 100], labels=['Bajo', 'Medio', 'Alto'], include_lowest=True)

    return d

df_score = calcular_score(df)
print(f"Score calculado para {len(df_score)} empleados")
print(f"Distribucion de niveles: ")
print(df_score['nivel_riesgo'].value_counts().sort_index())
```

```python
# PASO 1 — Seleccionar el top 10 y las columnas relevantes para el Director
# Ordenar descendente por score_riesgo y tomar los primeros 10


top10 = (df_score
         .sort_values('score_riesgo', ascending=False).head(10)[['EmployeeNumber', 'Department', 'JobRole', 'Age', 'YearsAtCompany', 'OverTime', 'score_riesgo', 'nivel_riesgo']].reset_index(drop=True))

print("Top 10 empleados en riesgo de Rotacion")
print(top10.to_string())
print("Wiiii")

```
