### Objetivo

Construir una app Shiny que permita explorar el dataset filtrando por
**rango de antigüedad** (`YearsAtCompany`) y que muestre:

1. Cuántos empleados tienen esa antigüedad

### Lo que necesitas implementar

| Elemento | Detalle |
|----------|---------|
| Input | `input_slider` de rango para `YearsAtCompany` (min=0, max=40) |

| Output 2 | Tabla con las primeras 10 filas filtradas |

### Pistas

- El slider de rango funciona igual que el de edad que ya usaste — solo cambia la columna
- Para calcular el porcentaje de rotación: `(d['Attrition'] == 'Yes').mean() * 100`

> No hay huecos — escribe el código completo en la Celda 3.


## Celda 1 — Exploración inicial del dataset

````python
# Instalación y exploración inicial
import pandas as pd
from shiny import App, ui, render

# Cargar el dataset
df = pd.read_csv("../hr_ibm.csv")

# Explorar la variable que usaremos como filtro

print("Exploracion de los años dentro de la compañia")
print(f"Rango: {df["YearsAtCompany"].min()} - {df["YearsAtCompany"].max()} años")
print(f"Promedio: {df["YearsAtCompany"].mean():.1f} años")
print(f"Mediana: {df["YearsAtCompany"].median():.0f} años")

print("Distribución de Tramos de Edad")
tramos = [0, 2, 5, 10, 20, 40]

for i in range(len(tramos)-1):
    sub = df[(df["YearsAtCompany"] >= tramos[i]) & (df["YearsAtCompany"] < tramos[i+1])]
    rot = (sub["Attrition"] == "Yes").mean()*100
    print(f" {tramos[i]:>2}-{tramos[i+1]:>2} años: {len(sub):>4} empleados | rotacion: {rot:.1f}%")
print()
print(f"Tasa de rotacion global: {(df['Attrition']=='Yes').mean():.1f}%")


````

## Celda 3 — Tu app completa

````python
# ── Interfaz de usuario ───────────────────────────────────────
app_ui = ui.page_fluid(
    ui.h2("Consulta de Antigüedad de Empleados "),
    ui.layout_sidebar(
        ui.sidebar(
            ui.input_slider(
                "antig",
                "Años en la empresa: ",
                min = 0, max = 40,
                value = [0, 40]
            )

        ),
        #outputs
        ui.output_text("resumen"),
        ui.output_table("tabla_resultado"),
    )
)
  

# ── Servidor ──────────────────────────────────────────────────
def server(input, output, session):

    def _filtrar():
        #el filtro del slider
        emin, emax = input.antig()
        return df[(df["YearsAtCompany"] >= emin) & (df["YearsAtCompany"] <= emax)]
    
    @render.text
    def resumen():
        d = _filtrar()
        n = len(d)

        if n == 0:
            return "Sin empleados en este rango"
        pct_rot = (d["Attrition"] == "Yes").mean()*100

        return (f"{n} empleados con {input.antig()[0]} - {input.antig()[1]} años " f"  Rotacion: {pct_rot:.1f}%")
    
    @render.table
    def tabla_resultado():
        cols = ["EmployeeNumber", "Department", "JobRole", "YearsAtCompany","Attrition", "MonthlyIncome"]
        return _filtrar()[cols].head(10)
    

app = App(app_ui, server)

#solo para verlo de forma local
import threading

def _iniciar():
    app.run(host="127.0.0.1", port=8000)

hilo = threading.Thread(target=_iniciar, daemon=True)
hilo.start()
print("App corriendo en local")
# Guardar como app.py → shiny run app.py

````
