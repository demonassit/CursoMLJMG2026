## Celda 1 — Instalación y carga del dataset

```python
# Instalar dependencias y cargar el dataset
import subprocess, sys 
subprocess.run([sys.executable, "-m", "pip", "install", "shiny", "pandas", "--quiet"])

import pandas as pd
from shiny import App, ui, render

# Cargar el dataset fuera de cualquier función (se ejecuta una sola vez)
df = pd.read_csv("../hr_ibm.csv")

#Construir listas de opciones para tres selectores 
opts_dept = ["Todos"] + sorted(df["Department"].unique().tolist())
opts_genero = ["Todos"] + sorted(df["Gender"].unique().tolist())
opts_edu = ["Todos"] + sorted(df["EducationField"].unique().tolist())


print(f"Datase: {df.shape[0]} filas X {df.shape[1]} columnas")
print(f"Generos Disponibles:   {df['Gender'].unique().tolist()} ")
print(f"Campos Educativos:   {df['EducationField'].unique().tolist()} ")
print(f"Salario Promedio Global:   {df['MonthlyIncome'].mean():,.0f} ")
# 'Todos' siempre es la primera opción — significa "sin filtro aplicado"

```

## Estructura de la app

La app tiene **tres selectores** y **dos outputs**:

| Control | Qué filtra |
|---------|-----------|
| Selector Departamento | Columna `Department` |
| Selector Género | Columna `Gender` |
| Selector Campo educativo | Columna `EducationField` |

| Output | Qué muestra |
|--------|------------|
| Texto | Cuántos empleados hay y cuál es su salario promedio |
| Tabla | Las primeras 10 filas del resultado |

> Los huecos están marcados con `# ← TU CÓDIGO`.
> Usa el selector de Departamento como modelo — los otros dos siguen el mismo patrón.


## Celda 3 — App con huecos para completar

```python
# ── Interfaz de usuario ───────────────────────────────────────
app_ui = ui.page_fluid(
    ui.h2("Explorador de empleados "),
    ui.layout_sidebar(
        ui.sidebar(
            ui.h4("Filtros"),
            # Selector 1 — Departamento (ya está completo, úsalo como modelo)
            ui.input_select("dept", "Departamento:", choices=opts_dept),
             # ═══════════════════════════════════════════════
            # HUECO 1: Agrega aquí el selector de Género
            # Usa el mismo patrón que el selector de dept
            # id sugerido: "genero"   choices: opts_genero
            # ═══════════════════════════════════════════════
            # ← TU CÓDIGO
            ui.input_select("genero", "Genero:", choices=opts_genero),


            # ═══════════════════════════════════════════════
            # HUECO 2: Agrega aquí el selector de Campo educativo
            # id sugerido: "edu"   choices: opts_edu
            # ═══════════════════════════════════════════════
            # ← TU CÓDIGO
            ui.input_select("edu", "Campo Educativo:", choices=opts_edu),

        ),
        #outputs
        ui.output_text("conteo_texto"),
        ui.output_table("tabla_resultado"),
    )
)
  

# ── Servidor ──────────────────────────────────────────────────
def server(input, output, session):

    def _filtrar():
        d = df.copy()

        # Filtro Departamento (ya está completo, úsalo como modelo)
        if input.dept() != "Todos":
            d = d[d["Department"] == input.dept()]


        # ═══════════════════════════════════════════════════
        # HUECO 3: Agrega los filtros de Género y EducationField
        # Sigue el mismo patrón que el filtro de dept
        # ═══════════════════════════════════════════════════
        # ← TU CÓDIGO (filtro género)
        # ← TU CÓDIGO (filtro campo educativo)
        if input.genero() != "Todos":
            d = d[d["Gender"] == input.genero()]
        
        if input.edu() != "Todos":
            d = d[d["EducationField"] == input.edu()]
        
        return d
    
    @render.text
    def conteo_texto():
        d = _filtrar()
        n = len(d)

        if n == 0:
            return "Sin resultados"

        sal = d["MonthlyIncome"].mean()
        return f"{n} empleados encontrados    -- Salario Promedio: ${sal:,.0f}"
    
    @render.table
    def tabla_resultado():
        cols = ["EmployeeNumber", "Department", "Gender", "EducationField", "JobRole", "Attrition", "MonthlyIncome"]
        return _filtrar()[cols].head(10)
    

app = App(app_ui, server)

#solo para verlo de forma local
import threading

def _iniciar():
    app.run(host="127.0.0.1", port=8000)

hilo = threading.Thread(target=_iniciar, daemon=True)
hilo.start()
print("App corriendo en local")
# Guardar como app.py → shiny run app.py
```
