## Celda 1 — Generar el dataset de ventas

````python
# Celda 1 — Generar el dataset de ventas
# Ejecuta esta celda para crear el dataset que usarás en la app
import pandas as pd
import numpy as np
import plotly.express as px
from shiny import App, ui, render, reactive

# Generar dataset reproducible de ventas retail
np.random.seed(42)
n=800
regiones = ["East", "West", "Central", "South"]
categorias = ["Furniture", "Office Supplies", "Technology"]

df_sales = pd.DataFrame(
    {
        "Region" : np.random.choice(regiones, n),
        "Category" : np.random.choice(categorias, n),
        "Month" : np.random.randint(1, 13, n),
        "Sales" : np.round(np.random.exponential(500, n) + 100, 2),
        "Profit" : np.round(np.random.normal(80, 150, n), 2),
        "Orders" : np.random.randint(1, 20, n)

    }
)
# Preparar opciones para los selectores
opts_region = ["Todos"] + sorted(df_sales["Region"].unique().tolist())
opts_category = ["Todos"] + sorted(df_sales["Category"].unique().tolist())

#verificar los datos
print(f"Dataset de ejemplo Generado: {df_sales.shape[0]} filas X {df_sales.shape[1]} columnas")
print()
print(df_sales.head(5).to_string(index=False))
print()
print(f"Ventas totales : ${df_sales['Sales'].sum():,.0f}")
print(f"Ganancia Total : ${df_sales['Profit'].sum():,.0f}")
print(f"Regiones : ${df_sales['Region'].unique().tolist()}")
print(f"Categorias : ${df_sales['Category'].unique().tolist()}")


````
