## Celda 1 — Instalación, importaciones y carga del dataset

```python
# ▶ EJECUTA — Celda 1: Instalación y carga del dataset de TechNova
# ──────────────────────────────────────────────────────────────────
# No modifiques esta celda. Ejecútala y verifica que ves:
# '✅ Dataset cargado: 1470 filas x 35 columnas'
#
# IMPORTANTE: el dataset se carga UNA SOLA VEZ aquí, fuera de
# cualquier función. Si lo pusiéramos dentro del servidor se
# recargaría con cada cambio de filtro — ineficiente en producción.
import pandas as pd
from shiny import App, ui, render


# Carga del dataset — fuera del server(), siempre
df = pd.read_csv("../hr_ibm.csv")

opts_dept = ["Todos"] + sorted(df["Departement"].unique().tolist())
opts_attr = ["Todos", "Yes", "No"]

##verificar que si salga :3 
```

## Celda 2 — Tu primera conexión reactiva

```python
# ▶ EJECUTA — Celda 2: Tu primera conexión reactiva
# ──────────────────────────────────────────────────────────────
# Guarda el código en app.py y ejecuta: shiny run app.py
# Observa cómo el texto cambia automáticamente al mover el selector.

app_ui_c2 = ui.page_fluid(
    ui.h2('Explorador de Departamentos'),
    ui.p('Cambia el selector -- El texto se actualiza automagicamente --'),
    ui.input_select(
        id='dept',
        label='Selecciona un departamento',
        choices=opts_dept
    ),
    ui.output_text('dept_texto')
)

def server_c2(input, output, session):
    @render.text
    def dept_texto():
        dept = input.dept()
        if dept == 'Todos':
            return f'Todos los departamentos: {len(df)} colaborales '
        filtrado = df[df["Departament"] == dept]
        rot = (filtrado["Attrition"] == 'Yes')
        return f'{dept}: {len{filtrado}} colaboradores | Rotaciones: {rot:.1f}%' 

app_c2 = App(app_ui_c2, server_c2)
# Guardar como app.py → shiny run app.py

##Aqui debemos incorporar como se visualizan esos datos de los cuadros de lista
```

## Celda 3 — Agrega el filtro de Attrition ✏️

```python
# ✏️ TU TURNO — Celda 3: Agrega el filtro de Attrition
# ──────────────────────────────────────────────────────────────
# La app ya tiene el selector de departamento y el slider de edad.
# Tu tarea: agregar el input_select de Attrition y el filtro
# correspondiente en _filtrar().
#
# Pistas:
# - En el ui: agrega ui.input_select('attrition', ...) en el sidebar
# - La columna del dataset se llama 'Attrition' (valores: 'Yes'/'No')
# - En _filtrar(): aplica el filtro con if input.attrition() != 'Todos'

app_ui_c3 = ui.page_fluid(
    ui.h2('Explorador de filtros multiples'),
    ui.layout_sidebar(
        ui.sidebar(
            ui.h4('Filtros del Director'),
            ui.input_select('dept', "Departamento:", choices=opts_dept),
            ui.input_select('attrition', "Rotación:", choices=opts_attr),
            ui.input_slider('edad', "Rango de Edad", min=18, max=60, value=[18, 60]),
        ),
        ui.output_text('conteo'),
        ui.output_table('tabla'),
    )

)

def server_c3(input, output, session):
    def _filtrar():
        d = df.copy()
        if input.dept() != 'Todos': d = d[d['Department'] == input.dept()]
        if input.attrition() != 'Todos': d = d[d['Attrition'] == input.attrition()]
        emin, emax = input.edad()
        return d[(d['Age'] >= emin) & (d['Age'] <= emax)]
    
    #Ahi les toca renderizar :3
```
