## Celda 3 — Crear `requirements.txt`

```python
# ▶ EJECUTA — Celda 3: Crear requirements.txt
# ──────────────────────────────────────────────────────────────
# Sin este archivo la publicación en shinyapps.io FALLARÁ.
# Ejecutar ANTES de intentar publicar.
import subprocess, sys

paquetes = ['shiny', 'pandas', 'plotly', 'rsconnect-python']

lineas = []

for pkg in paquetes:
    result = subprocess.run(
        [sys.executable, '-m', 'pip', 'show', pkg], capture_output=True, text=True
    )

    for line in result.stdout.split('\n'):
        if line.startswith('Version:'):
            version = line.split(': ')[1].strip()
            lineas.append(f'{pkg}>={version}')
            break

with open('requirements.txt', 'w') as f:
    f.write('\n'.join(lineas) + '\n')

print('documento de requisitos creado')
for l in lineas:
    print(f'        {l}')
print()
print("Acuerdense que este documento que se va a generar va dentro de la carpeta para el servidor")
```

## Celda 4 — Publicar en shinyapps.io

```python
# ▶ EJECUTA — Celda 4: Publicar en shinyapps.io
# ──────────────────────────────────────────────────────────────
# ESTOS COMANDOS SE EJECUTAN EN LA TERMINAL — no en Python.
# Token: shinyapps.io → Account → Tokens → Show → Copy

print('Comandos de terminal para publicar el dashboard:')
print()
print('# PASO 1: Configurar credenciales (una sola vez por computadora)')
print('rsconnect add \\')
print('  --account TU_NOMBRE_DE_CUENTA \\')
print('  --name mi-cuenta-shiny \\')
print('  --token TU_TOKEN \\')
print('  --secret TU_SECRET')
print()
print('# PASO 2: Publicar desde la carpeta del proyecto')
print('# Ejecutar desde la carpeta que contiene app.py y requirements.txt')
print('rsconnect deploy shiny . \\')
print('  --account TU_NOMBRE_DE_CUENTA \\')
print('  --name dashboard-technova-hr')
print()
print('# URL resultante:')
print('# https://TU_CUENTA.shinyapps.io/dashboard-technova-hr/')
print()
print('⚠️  Dar 48-72 horas para la publicación — el primer intento')
print('   puede tener errores de configuración.')
```
