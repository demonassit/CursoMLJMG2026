|---------|-------------|

## Celda 3 — Tu dashboard completo

````python
import pandas as pd
import numpy as np
import plotly.express as px
from shiny import App, ui, render, reactive

np.random.seed(7)
n=600
especialidades = ["Cardiology", "Orthopedics", "Neurology", "Oncology", "General"]
regiones =  ["Norte", "Sur", "Centro", "Oriente"]

df_hosp = pd.DataFrame(
    {
        "Speciality" : np.random.choice(especialidades, n, p=[.25, .2, .2, .15, .2]),
        "Region" : np.random.choice(regiones, n),
        "Month" : np.random.randint(1, 13, n),
        "Readmited" : np.random.choice(["Yes", "No"], n, p=[.2, .8]),
        "StayDays" : np.round(np.random.gamma(3,2,n) + 1, 1),
        "Satisfaction" : np.random.choice([1,2,3,4], n, p=[.1, .2, .45, .25]),
        "Age" : np.random.randint(18, 91, n),
        "Patients" : np.random.randint(5,50, n)
    }
)

opts_spec = ["Todos"] + sorted(df_hosp["Speciality"].unique().tolist())
opts_region = ["Todos"] + sorted(df_hosp["Region"].unique().tolist())

css = ui.tags.style("""body{ background:#FFFFFF; font-family:Segoe UI, Arial, sans-serif;} 
                    .navbar{ background-color:#1B1A20; }
                    .navbar-brand, .nav-link{ color:white; font-weight:500 }
                    .sidebar{ background:white; border-right:3px solid #388E3C; border-radius:8px; padding:16px; }
                    .kpi-card{ background:#AABBCC; border-radius:10px; padding:20px; text-aling:center; box-shadow:0 2px 8px rgba(0,0,0, 0.1);  }
                    .kpi-valor{ font-size:2em, font-weight:bold; }
                    .kpi-label{ font-size:0.85em, color:#666; margin-top:4px; } 
""")

app_ui = ui.page_navbar(
    ui.nav_panel("Resumen de Todo lo que Vimos", ui.layout_sidebar(
        ui.sidebar(
            ui.h5("Filtros"),
            ui.input_select("spec", "Especialidad", choices=opts_spec),
            ui.input_select("region", "Region", choices=opts_region),
            ui.input_slider("edad", "Rango de edad", min=18, max=90, value=[18,90]),
            ui.input_action_button("btn_reset", "Restablecer", class_="btn-sm mt-2")
        ),
        ui.h3("Indicadores clave"),
        ui.row(
            ui.column(4, ui.output_ui("kpi_readmision")),
            ui.column(4, ui.output_ui("kpi_estancia")),
            ui.column(4, ui.output_ui("kpi_satisfaccion"))
        ),
        ui.hr(),
        ui.h3("Visualizaciones"),
        ui.row(
            ui.column(6, ui.output_ui("grafica_readmin")),
            ui.column(6, ui.output_ui("grafica_scatter")),
        ),
    )),
    ui.nav_panel("Detalle de Pacientes", ui.output_ui("tabla_detalle")),
    title="Dashboard Hospitalario",
    header=css
)


def server(input, output, session):
    @reactive.calc
    def datos_filtrados():
        d = df_hosp()
        if input.spec() != "Todos": d = d[d["Speciality"] == input.spec()]
        if input.region() != "Todos": d = d[d["Region"] == input.region()]
        emin, emax = input.edad()
        return d[(d["Age"] >= emin) & (d["Age"] <= emax)]
    
    @reactive.effect
    @reactive.event(input.btn_reset)
    def _reset():
        ui.update_select("spec", selected="Todos")
        ui.update_select("region", selected="Todos")
        ui.update_slider("edad", value=[18,90])

    @render.ui
    def kpi_readmision():
        d = datos_filtrados()
        pct = (d["Readmitted"] == "Yes").mean()*100
        col = "#C62828" if pct>25 else("#F57F17" if pct>15 else "#2E7D32")
        return ui.div(ui.div(f"{pct:.1f}%", class_="kpi-valor", style=f"color:{col}"), ui.div("Tasa de readmisión < 30", class_="kpi-label"), class_="kpi-card")
    
    @render.ui
    def kpi_satisfacción():
        d = datos_filtrados()
        pct = (d["Satisfaction"]).mean()
        col = "#C62828" if pct>3.2 else("#F57F17" if pct>=2.5 else "#2E7D32")
        return ui.div(ui.div(f"{pct:.1f}%", class_="kpi-valor", style=f"color:{col}"), ui.div("Satisfacción del paciente: ", class_="kpi-label"), class_="kpi-card")
    
    @render.ui
    def kpi_estancia():
        d = datos_filtrados()
        pct = (d["StayDays"]).mean()
        col = "#C62828" if pct>8 else("#F57F17" if pct>5 else "#2E7D32")
        return ui.div(ui.div(f"{pct:.1f}%", class_="kpi-valor", style=f"color:{col}"), ui.div("Estancia promedio: ", class_="kpi-label"), class_="kpi-card")
    
    @render.ui
    def grafica_readmin():
        d = datos_filtrados()
        rot = d.groupby("Speciality")["Readmitted"].apply(lambda x:(x == "Yes").mean()*100).reset_index(name="Readmin_pct").round(1)
        fig = px.bar(rot, x="Speciality", y="Readmin_pct", color="Speciality", text_auto=".1f", title="Readmisión por especialidad (%)")
        fig.update_layout(showlegend=False, height=320)
        return ui.HTML(fig.to_html(include_plotlyjs="cdn"))
    
    @render.ui
    def grafica_scatter():
        d = datos_filtrados()
        fig = px.scatter(d, x="StayDays", y="Satisfaction", color="Readmitted", color_discrete_map={"Yes":"#E53935", "No":"#43A034"}, hover_data=["Speciality", "Age"], opacity=0.7, title="Estancia vs Satifacción")
        fig.update_layout(height=320)
        return ui.HTML(fig.to_html(include_plotlyjs="cdn"))
    
    @render.table
    def tabla_detalle():
        cols = ["Speciality", "Region", "Month", "Readmitted", "StayDays", "Satisfaction", "Age"]
        return datos_filtrados()[cols].head(20)
    
app = App(app_ui, server)

import threading

def _iniciar():
    app.run(host="127.0.0.1", port=8000)

hilo = threading.Thread(target=_iniciar, daemon=True)
hilo.start()
print("App corriendo en local")

#no lo voy a corregir porque es su tarea wiiiiiiiiiiiiiiiiiii

````
