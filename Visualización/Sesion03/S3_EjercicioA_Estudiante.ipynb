Agregar una **quinta tarjeta KPI** a la app del módulo que muestre el promedio de

- 1 = Bad, 2 = Good, 3 = Better, 4 = Best

### Tu tarea

| Parte | Qué hacer |
|-------|-----------|
| **CSS** | Cambiar el color de la navbar a uno de tu elección |

## Referencia: el KPI de rotación (úsalo como modelo)

@render.ui

## Celda 3 — App con huecos

````python
import pandas as pd
import plotly.express as px
from shiny import App, render, ui, reactive

df = pd.read_csv("../hr_ibm.csv")


opts_dept = ["Todos"] + sorted(df["Department"].unique().tolist())
opts_attr = ["Todos", "Yes", "No"]

css = ui.tags.style("""body{ background:#FFFFFF; font-family:Segoe UI, Arial, sans-serif;} 
                    .navbar{ background-color:#3C3E7A; }
                    .navbar-brand, .nav-link{ color:white; font-weight:500 }
                    .kpi-card{ border-radius:10px; padding:20px; text-aling:center; box-shadow:0 2px 8px rgba(0,0,0, 0.1);  }
                    .kpi-valor{ font-size:2em, font-weight:bold; }
                    .kpi-label{ font-size:0.85em, color:#666; margin-top:4px; }        

""")

# Font Awesome no viene incluido con Shiny: se carga desde el CDN para poder usar <i class="fa-solid ...">
fa_cdn = ui.tags.link(rel="stylesheet", href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.5.1/css/all.min.css")

app_ui = ui.page_navbar(
    ui.nav_panel(ui.tags.span(ui.tags.i(class_="fa-solid fa-circle-user", style="margin-right:6px;"), "Ejemplo de WorkLifeBalance"), ui.layout_sidebar(
        ui.sidebar(
            ui.h5("Filtros"),
            ui.input_select("dept", "Departamento", choices=opts_dept),
            ui.input_select("attrition", "Rotación: ", choices=opts_attr),
            ui.input_slider("edad", "Rango de Edad", min=18, max=60, value=[18,60]),
            ui.input_action_button("btn_reset", "Restablecer", class_="btn-sm mt-2")
        ),
        ui.h4("Indicadores clave"),
        ui.row(
            ui.column(2, ui.output_ui("kpi_total")),
            ui.column(2, ui.output_ui("kpi_rotación")),
            ui.column(3, ui.output_ui("kpi_salario")),
            ui.column(2, ui.output_ui("kpi_depto")),
            ui.column(3, ui.output_ui("kpi_balance"))
        ),
        ui.h4("Visualizaciones"),
        ui.row(
            ui.column(6, ui.output_ui("grafica_rotacion")),
            ui.column(6, ui.output_ui("grafica_scatter")),
        ),
    )),
    ui.nav_panel("Detalle", ui.output_ui("tabla_detalle")),
    title=ui.tags.span(ui.tags.i(class_="fa-solid fa-circle-user", style="margin-right:6px;"), "Dashboard de Analiticas de Balance"),
    header=ui.TagList(fa_cdn, css)
)

def server(input, output, session):
    @reactive.calc
    def datos_filtrados():
        d = df.copy()
        if input.dept() != "Todos": d = d[d["Department"] == input.dept()]
        if input.attrition() != "Todos": d = d[d["Attrition"] == input.attrition()]
        emin, emax = input.edad()
        return d[(d["Age"] >= emin) & (d["Age"] <= emax)]
    
    @reactive.effect
    @reactive.event(input.btn_reset)
    def _reset():
        ui.update_select("dept", selected="Todos")
        ui.update_select("attrition", selected="Todos")
        ui.update_slider("edad", value=[18,60])
    
    @render.ui
    def kpi_total():
        n = len(datos_filtrados())
        return ui.div(ui.div(f"{n:,}", class_="kpi-valor", style="color:#1F3864"), ui.div("Total empleados: ", class_="kpi-label"), class_="kpi-card")
    
    @render.ui
    def kpi_rotacion():
        d = datos_filtrados()
        pct = (d["Attrition"] == "Yes").mean()*100
        col = "#C62862" if pct > 20 else ("#F57F17" if pct > 10 else "#2E7D32")
        return ui.div(ui.div(f"{pct:.1f}%", class_="kpi-valor", style="color:{col}"), ui.div("Tasa de rotación: ", class_="kpi-label"), class_="kpi-card")
    
    @render.ui
    def kpi_salario():
        sal = datos_filtrados()["MonthlyIncome"].mean()
        return ui.div(ui.div(f"{sal:,.0f}", class_="kpi-valor", style="color:#2E7D32"), ui.div("Salario Promedio: ", class_="kpi-label"), class_="kpi-card")
    
    @render.ui
    def kpi_depto():
        d = datos_filtrados()
        if len(d) == 0:
            return ui.div(ui.div(f"--", class_="kpi-valor", style="color:#999"), ui.div("Mayor Rotación por Depto: ", class_="kpi-label"), class_="kpi-card")
        dep = d.groupby("Department")["Attrition"].apply(lambda x:(x == "Yes").mean()).idxmax()
        pct = d[d["Department"] == dep ]["Attrition"].eq("Yes").mean()*100
        return ui.div(ui.div(dep, class_="kpi-valor", style="color:#E67100"), ui.div(f"Mayor Rotación por Depto: {pct:.1f}%", class_="kpi-label"), class_="kpi-card") 
    
    @render.ui
    def kpi_balance():
        d = datos_filtrados()
        bal = d["WorkLifeBalance"].mean()
        col = "#2E7D32" if bal > 3.2 else("#F57F18" if bal >= 2.5 else "#C62828")
        return ui.div(ui.div(f"{bal:.2f}/4", class_="kpi-valor", style="color:{col}"), ui.div("Balance de vida - trabajo: ", class_="kpi-label"), class_="kpi-card")
    
    @render.ui
    def grafica_rotacion():
        d = datos_filtrados()
        rot = d.groupby("Department")["Attrition"].apply(lambda x:(x == "Yes").mean()*100).reset_index(name="Rotacion_pct").round(1)
        fig = px.bar(rot, x="Department", y="Rotacion_pct", color="Department", text_auto=".1f", title="Rotación por Departamento (%)")
        fig.update_layout(showlegend=False, height=320)
        return ui.HTML(fig.to_html(include_plotlyjs="cdn"))
    
    @render.ui
    def grafica_scatter():
        d = datos_filtrados()
        fig = px.scatter(d, x="MonthlyIncome", y="JobSatisfaction", color="Attrition", color_discrete_map={"Yes":"#E53935", "No":"#43A034"}, hover_data=["JobRole", "Age"], opacity=0.7, title="Salario vs Satifacción")
        fig.update_layout(height=320)
        return ui.HTML(fig.to_html(include_plotlyjs="cdn"))
    
    @render.table
    def tabla_detalle():
        cols = ["EmployeeNumber", "Department", "JobRole", "Age", "Attrition", "MonthlyIncome", "WorkLifeBalance"]
        return datos_filtrados()[cols].head(20)
    
app = App(app_ui, server)

import threading

def _iniciar():
    app.run(host="127.0.0.1", port=8000)

hilo = threading.Thread(target=_iniciar, daemon=True)
hilo.start()
print("App corriendo en local")

````
